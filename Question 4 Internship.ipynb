```python
from bs4 import BeautifulSoup 
import requests
import pandas as pd
```

```python
page=requests.get('https://shop.bewakoof.com/bestseller?sort=popular')
page
```

```python
soup= BeautifulSoup(page.content)
soup
```

```python
Name=[]
for i in soup.find_all('span',class_="sc-a6c4ca6a-0 PYPED"):
    Name.append(i.text)
```

```python
Name
```

```python
len(Name)
```

```python
NAME=Name[:10]
```

```python
price=[]
for q in soup.find_all('span',class_="sc-a6c4ca6a-0 ccQTLI sc-8020ee44-11 kbHGvj"):
    price.append(q.text)
```

```python
price
```

```python
images =[]
for w in soup.find_all("figure",class_="sc-8020ee44-0 eMFuRs"):
    url=w.img
    if url.has_attr('src'):
        images.append(url['src'])
    
    
    
images
```

```python
len(price)
```

```python
PRICE=price[:10]
```

```python
len(images)
```

```python
IMAGES=images[:10]
```

```python
df= pd.DataFrame({'Name':NAME,'Price':PRICE,'Image link':IMAGES})
```
