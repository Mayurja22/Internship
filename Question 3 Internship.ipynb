```python
soup= BeautifulSoup(page.content)
soup
```

```python
house_title=[]
for i in soup.find_all('h2',class_="heading-6 flex items-center font-semi-bold m-0"):
    house_title.append(i.text)
```

```python
house_title
```

```python
location=[]
for q in soup.find_all('div',class_="mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0.1p po:max-w-95"):
    location.append(q.text)
```

```python
location
```

```python
EMI = []
for w in soup.find_all('div',class_="font-semi-bold heading-6"):
    EMI.append(w.text)
```

```python
EMI
```

```python
import pandas as pd
```

```python
df= pd.DataFrame({'House Title':house_title,'Location Area':location})
```
