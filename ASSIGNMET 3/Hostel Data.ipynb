```python
import selenium 
import pandas as pd
from selenium import webdriver 
from selenium.common.exceptions import NoSuchElementException
import warnings 
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By 
import time 
import urllib.request
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.hostelworld.com/pwa/wds/s?q=London%2C+England&country=England&city=London&type=city&id=3&from=2024-09-19&to=2024-09-22&guests=2&page=1")
```

```python
SCROLL_PAUSE_TIME = 2
last_height = driver.execute_script("return document.documentElement.scrollHeight")

hostels = driver.find_elements(By.XPATH, "//div[@class='property-list']/div/div[6]/div")

hostel_data = []

for hostel in hostels:
    try:
    
        name = hostel.find_element(By.XPATH, "//div[@class='property-name']/span").text.strip()

        distance = hostel.find_element(By.XPATH, "//div[@class='property-distance']/span[2]").text.strip()

      
        try:
            rating = hostel.find_element(By.XPATH, "//div[@class='rating medium theme-null']/span[1]").text.strip()
        except:
            rating = 'No Rating'

        
        try:
            total_reviews = hostel.find_element(By.XPATH, "//div[@class='rating medium theme-null']/span[3]").text.split()[0].strip()
        except:
            total_reviews = 'No Reviews'

       
        try:
            overall_reviews = hostel.find_element(By.XPATH, "//div[@class='rating medium theme-null']/span[2]").text.strip()
        except:
            overall_reviews = 'No Overall Review'

        
        try:
            private_price = hostel.find_element(By.XPATH, "//div[contains(@class, 'accommodation-price')]//strong[@class='current']").text.strip()
        except:
            private_price = 'N/A'

        try:
            dorm_price = hostel.find_element(By.XPATH, "//div[@class='property-accommodation-prices']/div[2]//div[contains(@class, 'accommodation-price')]//strong[@class='current']").text.strip()
        except:
            dorm_price = 'N/A'

        hostel_url = hostel.find_element(By.TAG_NAME, 'a').get_attribute('href')
        driver.execute_script("window.open(arguments[0]);", hostel_url)
        driver.switch_to.window(driver.window_handles[1])  # Switch to the new tab

        
        time.sleep(3)

        
        try:
            facilities = [fac.text.strip() for fac in driver.find_elements(By.XPATH, '//span[@class="facility-name body-2-reg"]')]
        except:
            facilities = []

       
        try:
            description = driver.find_element(By.XPATH, "//div[@class='about-description']/p").text.strip()
        except:
            description = 'No Description'

       
        driver.close()
        driver.switch_to.window(driver.window_handles[0])

      
        hostel_data.append({
            "Hostel Name": name,
            "Distance from City Centre": distance,
            "Rating": rating,
            "Total Reviews": total_reviews,
            "Overall Review": overall_reviews,
            "Privates from Price": private_price,
            "Dorms from Price": dorm_price,
            "Facilities": facilities,
            "Property Description": description
        })

    except Exception as e:
        print(f"Error extracting hostel data: {e}")


try:
    next_button = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/div[2]/div[2]/div[1]/div/div[7]/div/button[2]')
    next_button.click()

    
    time.sleep(5)

except:
    print("No more pages available. Scraping complete.")
    


for hostel in hostel_data:
    print(f"Hostel Name: {hostel['Hostel Name']}")
    print(f"Distance from City Centre: {hostel['Distance from City Centre']}")
    print(f"Rating: {hostel['Rating']}")
    print(f"Total Reviews: {hostel['Total Reviews']}")
    print(f"Overall Review: {hostel['Overall Review']}")
    print(f"Privates from Price: {hostel['Privates from Price']}")
    print(f"Dorms from Price: {hostel['Dorms from Price']}")
    print(f"Facilities: {', '.join(hostel['Facilities'])}")
    print(f"Property Description: {hostel['Property Description']}\n")
```
